# Image Manipulation 

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image_path = 'data/IM-0117-0001.jpeg'
```

###  Load and display a zero-channel image 

```python
img = cv2.imread(image_path,0)
# img = cv2.imread(image_path,1)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img.shape, img.dtype
```

# Method 1 - Using OpenCV

```python
new_img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
print("New shape:", new_img.shape)
plt.imshow(new_img)
```

# Method 2 - Using np.stack

```python
stacked_img = np.stack((img,)*3, -1)
print("Stacked image shape:", stacked_img.shape)
plt.imshow(stacked_img)
```

# Pixels difference between two methods

```python
print("Pixels difference:", np.sum(stacked_img - new_img))
```
